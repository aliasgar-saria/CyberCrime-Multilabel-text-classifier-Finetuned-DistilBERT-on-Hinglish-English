```python
import pandas as pd

# Read the file
df = pd.read_csv('categorized_crimes_complete.csv')

# Check unique values and nulls
category_unique_count = df['category'].nunique()
sub_category_unique_count = df['sub_category'].nunique()
print(f"Unique categories: {category_unique_count}")
print(f"Unique sub-categories: {sub_category_unique_count}")

# Null value checks
null_in_category = df['category'].isnull().sum()
null_in_subcategory = df['sub_category'].isnull().sum()
print(f"\nNull values in category: {null_in_category}")
print(f"Null values in sub_category: {null_in_subcategory}")

# Total nulls
total_null = df.isnull().sum().sum()
print(f"\nTotal null values in dataframe: {total_null}")
rows_with_nulls = df[df.isnull().any(axis=1)].shape[0]
print(f"Number of rows containing null values: {rows_with_nulls}")

# Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15, 10))

# Plot 1: Category value counts
plt.subplot(2, 2, 1)
df['category'].value_counts().plot(kind='bar')
plt.title('Category Distribution')
plt.xticks(rotation=45)

# Plot 2: Sub-category value counts
plt.subplot(2, 2, 2)
df['sub_category'].value_counts().plot(kind='bar')
plt.title('Sub-category Distribution')
plt.xticks(rotation=45)

# Plot 3: Null values heatmap
plt.subplot(2, 2, 3)
sns.heatmap(df[['category', 'sub_category']].isnull(), cmap='viridis', yticklabels=False)
plt.title('Null Values Heatmap')

# Plot 4: Category vs Sub-category count
plt.subplot(2, 2, 4)
sns.countplot(data=df, x='category', hue='sub_category')
plt.title('Category vs Sub-category Distribution')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Additional pie charts
plt.figure(figsize=(15, 5))

# Category pie chart
plt.subplot(1, 2, 1)
df['category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Category Distribution (%)')

# Sub-category
```

```python
# Get unique categories and subcategories
unique_categories = sorted(df['category'].unique())
unique_subcategories = sorted(df['sub_category'].unique())

print("Unique Categories:")
for idx, category in enumerate(unique_categories, 1):
    print(f"{idx}. {category}")

print("\nUnique Subcategories:")
for idx, subcategory in enumerate(unique_subcategories, 1):
    print(f"{idx}. {subcategory}")

# For a more detailed view, show category-subcategory relationships
print("\nCategory-Subcategory Relationships:")
for category in unique_categories:
    subcats = sorted(df[df['category'] == category]['sub_category'].unique())
    print(f"\n{category}:")
    for subcat in subcats:
        print(f"  - {subcat}")
```

Renaming 3 subcategories of csv 'Sexually Explicit Act
  - Sexually Obscene Material
  - Sexually Obscene material' into 'Sexual Harassment'

```python
# Create a mapping dictionary for the replacements
replacement_dict = {
    'Sexually Explicit Act': 'Sexual Harassment',
    'Sexually Obscene Material': 'Sexual Harassment',
    'Sexually Obscene material': 'Sexual Harassment'
}

# Replace the values in the sub_category column
df['sub_category'] = df['sub_category'].replace(replacement_dict)

# Verify the changes
print("Updated Category-Subcategory Relationships:")
unique_categories = sorted(df['category'].unique())
for category in unique_categories:
    subcats = sorted(df[df['category'] == category]['sub_category'].unique())
    print(f"\n{category}:")
    for subcat in subcats:
        print(f"  - {subcat}")
```

```python
# Save the updated DataFrame to CSV
df.to_csv('categorized_crimes_complete.csv', index=False)
print("File saved successfully as 'categorized_crimes_complete.csv'")

# Verify the save by reading the first few rows
print("\nVerifying saved data:")
verification_df = pd.read_csv('categorized_crimes_complete.csv')
print("\nUnique subcategories in saved file:")
print(sorted(verification_df['sub_category'].unique()))
```
